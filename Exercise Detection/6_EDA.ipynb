## Model Comparison

### Importing the Hyperparameter-tuned datasets

```python
# Importing all the necessary libraries/modules
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# importing the datasets
df_wPCA = pd.read_excel("Accuracy_wPCA.xlsx")
df_woPCA = pd.read_excel("Accuracy_woPCA.xlsx")
```

```python
df_wPCA
```

```python
df_woPCA
```

### Dataframe with PCA

##### Algorhythm vs Accuracy Score

```python
sns.catplot(data = df_wPCA, x = 'Algorhythm', y = 'Accuracy Score', kind = 'box')
plt.xticks(rotation=90)
```

*Inference:* LogisticRegression has an overall better performance when compared to other models.

##### n_components vs Accuracy Score

```python
fig, ax = plt.subplots(figsize=(15,8))
sns.pointplot(data = df_wPCA, y = 'Accuracy Score', x = 'n_components', hue = 'Algorhythm', ax = ax)
plt.xticks(rotation=90)
plt.show()
```

*Inference:* RandomForestClassifier and LogisticRegression have higher accuracies than the other 2 models. The increase in number of components does not indicate any fluctuation in the accuracy.

##### computation vs Accurcay Score

```python
fig, ax = plt.subplots(figsize=(15,8))
sns.scatterplot(data = df_wPCA, y = 'Accuracy Score', x = 'Computation Time', hue = 'Algorhythm', ax = ax)
plt.xticks(rotation=90)
plt.show()
```

### Dataframe without PCA

##### Algorhythm vs Accuracy Score

```python
sns.catplot(data = df_woPCA, x = 'Algorhythm', y = 'Accuracy Score', kind = 'box')
plt.xticks(rotation=90)
```

*Inference:* LogisticRegression and RandomForestClassifier has an overall better performance when compared to other models and both the models share the same highest accuracy.

##### computation vs Accurcay Score

```python
fig, ax = plt.subplots(figsize=(15,8))
sns.scatterplot(data = df_woPCA, y = 'Accuracy Score', x = 'Computation Time', hue = 'Algorhythm', ax = ax)
plt.xticks(rotation=90)
plt.show()
```
